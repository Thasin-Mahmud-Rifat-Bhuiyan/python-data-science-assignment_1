# Question 5:

```python
# Function to calculate average accuracy
def calculate_experiment_accuracy(a, b, c):

    print("Experiment Accuracies:", a, b, c)  #Experiment Accuracies
    avg = (a + b + c) / 3  #Average accuracy
    return avg  # Return the result


# Call the function
result=calculate_experiment_accuracy(89, 91.5, 93.5)

print("Average Accuracy =", result)
```
